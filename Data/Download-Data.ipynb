***Récupère les données voulu et les sauvegardes dans un fichier .parquet***

```python
import vectorbt as vbt
import pandas as pd

symbol = ['BROCCOLI714USDT']

time = "1m"

# Téléchargement des données
data = vbt.BinanceData.download(
    symbol, 
    start='2025-10-07',
    #end='2022-01-01',
    interval=time,
    tz_convert='Europe/Paris',
)
```

```python
data.stats()
len(data.get())
```

```python
# Sauvegarde des données
df = data.get()
df.to_parquet('BROCCOLI714USDT_100d_2025-10-01_TO_2026-01-07_1m',engine='pyarrow', index=True)
```
