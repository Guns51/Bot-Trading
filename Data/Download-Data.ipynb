***Récupère les données voulu et les sauvegardes dans un fichier .parquet***

```python
import vectorbt as vbt
import pandas as pd

symbol = ['ETHUSDT']

time = "1m"

# Téléchargement des données
data = vbt.BinanceData.download(
    symbol, 
    start='2026-01-04',
    #end='2022-01-01',
    interval=time,
    tz_convert='Europe/Paris',
)
```

```python
data.stats()
len(data.get())
```

```python
# Sauvegarde des données
df = data.get()
df.to_parquet('ETHUSDT_1.8d_2026-01-04_TO_2026-01-04_1m',engine='pyarrow', index=True)
```
